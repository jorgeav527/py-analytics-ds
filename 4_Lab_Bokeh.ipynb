````python
#EJEMPLO COMBINADO
import numpy as np

from bokeh.io import output_notebook
from bokeh.plotting import figure, show

# data
day_num = np.linspace(1, 10, 10)
daily_words = [450, 628, 488, 210, 287, 791, 508, 639, 397, 943]
cumulative_words = np.cumsum(daily_words)

output_notebook()

# figura con eje tipo datetime
fig = figure(title='Progreso',
             height=400, width=700,
             x_axis_label='Día', y_axis_label='Palabras',
             x_minor_ticks=2, y_range=(0, 6000),
             toolbar_location=None)

# representar palabras como columnas
fig.vbar(x=day_num, bottom=0, top=daily_words, 
         color='blue', width=0.75, 
         legend_label='Diario')

# suma como línea
fig.line(x=day_num, y=cumulative_words, 
         color='gray', line_width=1,
         legend_label='Acumulado')

# ubicación de la leyenda
fig.legend.location = 'top_left'

show(fig)
````

````python
import seaborn as sns

from bokeh.plotting import figure, show, gridplot
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.io import output_notebook

"""
💡 La Analogía del "Diccionario Mágico"
Técnicamente, el ColumnDataSource convierte tu tabla en un diccionario de listas.
Si tu tabla original es:
| name | mpg | weight |
| :--- | :--- | :--- |
| chevrolet | 18 | 3504 |
| mazda | 24 | 2150 |
El ColumnDataSource lo ve así:
'name': ['chevrolet', 'mazda']
'mpg': [18, 24]
'weight': [3504, 2150]
"""

# Esto es para que el gráfico se vea dentro de tu Jupyter Notebook
output_notebook()
````

````python
# 1. Cargamos la data original
df = sns.load_dataset('mpg')
````

````python
# 2. FILTRAMO CON PANDAS (como ya sabes hacerlo)
df_europa = df[df['origin'] == 'europe'].copy()

# 3. CONVERTIMOS EL RESULTADO A ColumnDataSource
source = ColumnDataSource(df_europa)

# 4. GRAFICAMOS
p = figure(title="Autos Europeos: Peso vs MPG", width=600, height=400)
p.scatter('weight', 'mpg', size=10, source=source, marker="circle",
          color="navy", alpha=0.5, hover_fill_color="firebrick")

# 5. CONFIGURAMOS EL HOVER (El "cartelito")
hover = HoverTool()
hover.tooltips = [
    ("Modelo", "@name"),       # @ indica que busque en la columna 'name' del source
    ("Consumo (MPG)", "@mpg"),
    ("Caballos", "@horsepower")
]

# 5. Agregamos la herramienta al gráfico
p.add_tools(hover)

show(p)
````

### Linked Plots (Gráficos Vinculados)

"¿Un auto que tiene muchos caballos de fuerza (horsepower) siempre es más pesado (weight), o existen autos ligeros con motores muy potentes? Vamos a crear dos gráficos: uno de Potencia y otro de Peso. Si seleccionas los autos más potentes en el primer gráfico, ¿se iluminarán también como los más pesados en el segundo? ¡Descubrámoslo!"

````python
# 2. Configuración de herramientas (Agregamos 'lasso_select' para seleccionar puntos)
tools = "lasso_select,box_select,wheel_zoom,reset,save"

# 3. PRIMER GRÁFICO: Caballos de Fuerza vs MPG
p1 = figure(title="Potencia vs Consumo", x_axis_label="Horsepower", 
            y_axis_label="MPG", width=400, height=400, tools=tools)

p1.scatter(x='horsepower', y='mpg', source=source, size=8, 
           color="navy", alpha=0.4, selection_color="orange")

# 4. SEGUNDO GRÁFICO: Peso vs MPG (Vinculado automáticamente por el 'source')
p2 = figure(title="Peso vs Consumo", x_axis_label="Weight", 
            y_axis_label="MPG", width=400, height=400, tools=tools)

# Vinculamos los rangos de las Y para que si mueves uno, se mueva el otro (opcional)
p2.y_range = p1.y_range

p2.scatter(x='weight', y='mpg', source=source, size=8, 
           color="firebrick", alpha=0.4, selection_color="orange")

# 5. Organizamos en una rejilla (Grid)
layout = gridplot([[p1, p2]])

show(layout)
````

### Diagrama de Dispersión Interactivo con Filtro por Leyenda

"¿Cómo influye el país de origen en la relación entre el peso de un auto y su eficiencia? ¿Existen autos pesados que sean eficientes, o es una característica exclusiva de ciertas regiones?"

````python
p = figure(title="Haz clic en la leyenda para filtrar", width=700, height=450,
           tools="pan,wheel_zoom,reset,hover")

# Graficamos cada origen por separado para que sean interactivos en la leyenda
for origen, color in zip(['usa', 'japan', 'europe'], ['navy', 'firebrick', 'forestgreen']):
    source = ColumnDataSource(df[df['origin'] == origen])
    p.scatter('weight', 'mpg', source=source, size=10, color=color, 
              alpha=0.6, legend_label=origen)

# EL TRUCO MÁGICO:
p.legend.click_policy = "hide" # Al hacer clic en la leyenda, el grupo se oculta
p.legend.location = "top_right"

show(p)
````

### Mapas Geográficos (GMapPlot / Tile Providers)

"¿Dónde se fabrican los autos que estamos analizando? Ubiquemos en el mapa las tres grandes regiones (USA, Europa y Japón) para visualizar su importancia en el mercado."

````python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
# Eliminamos la línea que daba error y usamos esta:
from bokeh.io import output_notebook

output_notebook()

# 1. Coordenadas aproximadas (Mercator)
datos = {
    'coords_x': [-10000000, 2000000, 15000000], 
    'coords_y': [4500000, 6000000, 4500000],
    'region': ['USA', 'Europa', 'Japón']
}
source = ColumnDataSource(datos)

# 2. Crear la figura
p = figure(x_range=(-15000000, 20000000), y_range=(-2000000, 9000000),
           x_axis_type="mercator", y_axis_type="mercator",
           title="Sedes de Fabricación (Versión 2026)",
           width=800, height=450)

# 3. AGREGAMOS EL MAPA (La forma nueva y fácil)
# "CartoDB Positron" es el nombre del estilo de mapa
p.add_tile("CartoDB Positron") 

# 4. Dibujamos los puntos
p.scatter(x='coords_x', y='coords_y', size=20, color="firebrick", alpha=0.7, source=source)

show(p)
````

````python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, Select, CustomJS
from bokeh.layouts import column
from bokeh.io import output_notebook

output_notebook()

# --- Datos por región ---
data_dict = {
    "Todos": {
        "coords_x": [-10000000, 2000000, 15000000],
        "coords_y": [4500000, 6000000, 4500000],
        "region": ["USA", "Europa", "Japón"]
    },
    "USA": {
        "coords_x": [-10000000],
        "coords_y": [4500000],
        "region": ["USA"]
    },
    "Europa": {
        "coords_x": [2000000],
        "coords_y": [6000000],
        "region": ["Europa"]
    },
    "Japón": {
        "coords_x": [15000000],
        "coords_y": [4500000],
        "region": ["Japón"]
    },
}

# Fuente inicial (todos los puntos)
source = ColumnDataSource(data_dict["Todos"])

# Figura con mapa
p = figure(x_range=(-15000000, 20000000), y_range=(-2000000, 9000000),
           x_axis_type="mercator", y_axis_type="mercator",
           title="Sedes de Fabricación (Dinámico)",
           width=800, height=450)

p.add_tile("CartoDB Positron")

p.scatter(x='coords_x', y='coords_y', size=20, color="firebrick",
          alpha=0.7, source=source)

# Dropdown
select = Select(title="Filtrar por región:", value="Todos",
                options=list(data_dict.keys()))

# Callback JavaScript (no necesita servidor)
select.js_on_change("value", CustomJS(args=dict(src=source, data=data_dict), code="""
    const region = cb_obj.value;
    src.data = {
        coords_x: data[region].coords_x,
        coords_y: data[region].coords_y,
        region: data[region].region
    };
    src.change.emit();
"""))

# Layout dinámico
show(column(select, p))
````

### Gráfico de Dispersión con Filtro Dinámico en el Cliente

"¿A partir de qué nivel de potencia (horsepower) el ahorro de combustible deja de ser una opción técnica posible?"

````python
import seaborn as sns
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, CustomJS, Slider
from bokeh.layouts import column
from bokeh.io import output_notebook

# 1. Preparar datos
df = sns.load_dataset('mpg')
source = ColumnDataSource(df) # El que se mueve
original_source = ColumnDataSource(df) # El respaldo con todos los datos

output_notebook()

# 2. Crear el gráfico
p = figure(title="Filtro de Potencia (Horsepower)", width=600, height=400,
           x_range=(40, 250), y_range=(5, 50))

p.scatter('horsepower', 'mpg', source=source, size=10, color="navy", alpha=0.5)

# 3. CREAR EL SLIDER DE BOKEH (No de Python/Jupyter)
# Este slider sí aparecerá porque es parte del HTML del gráfico
slider = Slider(start=45, end=230, value=45, step=1, title="Potencia Mínima")

# 4. El "Cerebro" (JavaScript) para que funcione sin recargar la página
callback = CustomJS(args=dict(source=source, original_source=original_source, slider=slider), code="""
    const data = source.data;
    const original_data = original_source.data;
    const val = slider.value;
    
    // Limpiamos los datos actuales
    for (let key in original_data) {
        data[key] = [];
        // Filtramos: solo agregamos los que superan el valor del slider
        for (let i = 0; i < original_data['horsepower'].length; i++) {
            if (original_data['horsepower'][i] >= val) {
                data[key].push(original_data[key][i]);
            }
        }
    }
    source.change.emit();
""")

slider.js_on_change('value', callback)

# 5. Mostrar todo junto
layout = column(slider, p)
show(layout)
````

from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route("/")
def index():
    # Create a very simple plot
    p = figure(title="Very Simple Bokeh Plot", height=300, width=500)
    p.line([1, 2, 3, 4], [4, 7, 2, 6], line_width=3)

    # Convert Bokeh plot to script + div
    script, div = components(p)

    return render_template("index.html", script=script, div=div)

if __name__ == "__main__":
    app.run(debug=True)
