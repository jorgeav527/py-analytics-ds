```python
from datetime import date

import matplotlib.pyplot as plt
import pandas as pd
import streamlit as st

# --- CONFIGURACIÓN DE PÁGINA ---
st.set_page_config(page_title="Cibertec-208", layout="wide", page_icon="📊")

# --- ESTILOS PERSONALIZADOS (Opcional para ese toque "Pro") ---
st.markdown(
    """
    <style>
    /* Usamos un color que se adapte o simplemente un borde */
    [data-testid="stMetric"] {
        background-color: rgba(28, 131, 225, 0.1); /* Un azul muy transparente */
        padding: 15px;
        border-radius: 10px;
        border: 1px solid rgba(28, 131, 225, 0.3);
    }
    </style>
    """,
    unsafe_allow_html=True,
)

# ==========================================
# MÓDULO 1: PRESENTACIÓN Y MULTIMEDIA
# ==========================================
st.title("🚀 Basic - class: Streamlit para Data Analytics")
st.divider()

tab_intro, tab_media = st.tabs(["📝 Documentación", "🖼️ Multimedia"])

with tab_intro:
    col_t1, col_t2 = st.columns([2, 1])
    with col_t1:
        st.header("Componentes de Texto")
        st.write(
            "Streamlit permite usar Markdown, lo que facilita crear reportes dinámicos."
        )
        st.code(
            """
# Función para normalizar nombres de columnas
def clean_data(df):
    df.columns = [col.lower().replace(" ", "_") for col in df.columns]
    return df.dropna()

df_clean = clean_data(raw_data)
""",
            language="python",
            line_numbers=True,
        )
        st.json({"proyecto": "Streamlit Lab", "estado": "Activo", "version": 1.0})
        st.info("Tip: Usa `st.divider()` para separar secciones visualmente.")
        st.success("✅ **Éxito:** El dataset se cargó correctamente.")
        st.info("ℹ️ **Información:** Este gráfico muestra la tendencia anual.")
        st.warning("⚠️ **Advertencia:** Faltan valores nulos en la columna 'Edad'.")
        st.error("🚨 **Error:** No se pudo conectar a la base de datos.")

    with col_t2:
        st.subheader("Metricas")
        st.metric(label="Versión", value="1.0.2", delta="Estable")
        st.metric(label="Registros Activos", value="1,250", delta="+12% vs ayer")
        st.metric(
            label="Tiempo de Carga", value="0.8s", delta="-0.2s", delta_color="inverse"
        )

with tab_media:
    st.subheader("Recursos Externos")
    with st.expander(
        "🎵 Reproducir Audio de Muestra (Análisis de Señal)", expanded=True
    ):
        st.audio(
            "https://www.soundhelix.com/examples/mp3/SoundHelix-Song-1.mp3",
            format="audio/mp3",
        )
    st.divider()
    c1, c2 = st.columns(2)
    with c1:
        st.image(
            "https://images.unsplash.com/photo-1518770660439-4636190af475",
            caption="Hardware Analysis",
            use_container_width=True,
        )
    with c2:
        st.video("https://www.youtube.com/watch?v=dQw4w9WgXcQ")

# ==========================================
# MÓDULO 2: INTERACTIVIDAD Y CAPTURA
# ==========================================
# --- 1. ENTRADA DE DATOS (SIDEBAR) ---
with st.sidebar:
    st.header("⌨️ Configuración")
    user_name = st.text_input("Analista:", placeholder="Nombre")

    # Agrupamos los widgets esenciales
    edad = st.number_input("Edad", 18, 100, 31)
    nivel_py = st.select_slider(
        "Nivel Python", ["Novato", "Intermedio", "Avanzado", "Ninja"], value="Avanzado"
    )
    lenguajes = st.multiselect(
        "Lenguajes que dominas:",
        ["Python", "R", "SQL", "Julia", "Scala", "C++", "Rust", "Go"],
        default=["Python"],  # Opción preseleccionada
    )
    nivel = st.select_slider(
        "Nivel Técnico",
        ["Universidad", "Maestria", "Doctorado", "Ninja"],
        value="Universidad",
    )
    color_fav = st.radio("Color Favorito", ["Rojo", "Verde", "Azul"], horizontal=True)
    fecha_analisis = st.date_input("Fecha del Análisis", date.today())
    notas = st.text_area(
        "Notas Adicionales", placeholder="Escribe observaciones aquí..."
    )
    acepto = st.checkbox("Certificar información", value=True)

    # Botón disparador
    btn_generar = st.button("🚀 Generar Resumen", use_container_width=True)

st.header("🕹️ Centro de Control de Usuario")
if btn_generar:
    if not user_name:
        st.error("⚠️ Debes ingresar un nombre en el panel lateral.")
    else:
        st.balloons()

        # Resumen limpio y minimalista dentro de un Expander
        with st.expander("📄 EXPEDIENTE DEL ANALISTA", expanded=True):
            st.markdown(f"### ANALISTA: {user_name.upper()}")

            # Fila 1: Datos Técnicos
            c1, c2, c3 = st.columns(3)
            c1.write(f"**Nivel:** {nivel}")
            c2.write(f"**Stack:** {lenguajes}")
            c3.write(f"**Edad:** {edad} años")

            # Fila 2: Fechas y Archivos
            d1, d2, d3 = st.columns(3)
            d1.write(f"**Fecha:** {fecha_analisis.strftime('%d/%m/%Y')}")
            d2.write(f"**Nivel:** {nivel}")
            d3.write(f"**Color Elegido:** {color_fav}")

            st.divider()

            # Mostrar notas solo si existen
            if notas:
                st.markdown("**Observaciones:**")
                st.info(notas)

            st.success("Informe generado y certificado exitosamente.")

else:
    st.info("👈 Configura todos los parámetros en el Sidebar para ver el resultado.")

st.divider()

# ==========================================
# MÓDULO 3: Data
# ==========================================
st.header("🔍 Explorador de Datos (Gapminder)")
try:
    df = pd.read_csv("resources/gapminder.csv")
    search_query = st.text_input(
        "🔎 Buscar país, año o continente:",
        placeholder="Ej: Mexico, 2007, Asia...",
        help="Escribe para filtrar la tabla automáticamente",
    )
    if search_query:
        # Buscamos en todo el dataframe
        mask = (
            df.astype(str)
            .apply(lambda x: x.str.contains(search_query, case=False))
            .any(axis=1)
        )
        df_filtrado = df[mask]
    else:
        df_filtrado = df

    total_filas = len(df)
    filas_encontradas = len(df_filtrado)
    porcentaje = filas_encontradas / total_filas

    st.write(
        f"Resultados encontrados: **{filas_encontradas}** de **{total_filas}** registros"
    )
    st.progress(porcentaje)

    st.dataframe(df_filtrado, use_container_width=True, height=350)

    with st.expander("⚙️ Descripción", expanded=False):
        st.dataframe(df_filtrado.describe())

    with st.expander("⚙️ Info", expanded=False):
        # Truco pro: Capturamos el buffer de df.info() para mostrarlo en Streamlit
        import io

        buffer = io.StringIO()
        df_filtrado.info(buf=buffer)
        info_text = buffer.getvalue()
        st.text(info_text)
except:
    st.warning("⚠️ No se encontró 'gapminder.csv'. Por favor, sube un archivo abajo.")

# ==========================================
# MÓDULO 4: ANÁLISIS PROFUNDO Y CARGA
# ==========================================
st.header("🧪 Análisis Avanzado")

uploaded_file = st.file_uploader("Sube tu propio dataset (CSV)", type="csv")

if uploaded_file:
    data = pd.read_csv(uploaded_file)
    st.success("Archivo cargado!")

    # Análisis dinámico con Matplotlib
    st.subheader("Evolución de Esperanza de Vida")
    paises = sorted(data["country"].unique())
    pais_sel = st.selectbox("Selecciona un país para analizar:", paises)

    df_filtrado = data[data["country"] == pais_sel]

    fig, ax = plt.subplots()
    ax.plot(df_filtrado["year"], df_filtrado["lifeExp"], marker="o", color="teal")
    ax.set_title(f"Tendencia en {pais_sel}")
    st.pyplot(fig)
else:
    st.info("Sube un archivo para habilitar el análisis dinámico.")
```
