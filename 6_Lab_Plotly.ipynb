```python
import plotly.express as px # Express: Es rápido y hace muchas suposiciones por ti (la "cámara automática").
import pandas as pd

data = {
    'Día': ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo'],
    'Tazas_Cafe': [2, 4, 8, 12, 5, 1, 0],
    'Productividad': [60, 85, 95, 10, 50, 20, 5],
    'Estado_Mental': ['Normal', 'Enfocado', 'Super sayayin', 'Bug viviente', 'Cansado', 'Modo Zen', 'Dormido']
}

df_fun = pd.DataFrame(data)

# Creamos el gráfico de líneas
fig = px.line(df_fun, 
              x='Día', 
              y='Productividad', 
              title='Relación entre Café y Productividad del Alumno ☕🚀',
              text='Estado_Mental', # Añade etiquetas sobre los puntos
              markers=True)

# Personalización para que se vea genial
fig.update_traces(textposition="top center", line_color='#FF5733')
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# Usamos los mismos datos que creamos antes
data = {
    'Día': ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo'],
    'Tazas_Cafe': [2, 4, 8, 12, 5, 1, 0],
    'Productividad': [60, 85, 95, 10, 50, 20, 5],
    'Estado_Mental': ['Normal', 'Enfocado', 'Super sayayin', 'Bug viviente', 'Cansado', 'Modo Zen', 'Dormido']
}
df_fun = pd.DataFrame(data)

# 1. Inicializamos la "Figura" (el lienzo vacío)
fig = go.Figure()

# 2. Añadimos el "Trace" (el trazo o la línea)
fig.add_trace(go.Scatter(
    x=df_fun['Día'], 
    y=df_fun['Productividad'],
    mode='lines+markers+text', # Definimos que queremos líneas, puntos y texto
    name='Productividad',
    text=df_fun['Estado_Mental'],
    textposition="top center",
    line=dict(color='firebrick', width=4),
    marker=dict(size=12, symbol='star')
))

# 3. Personalizamos el diseño (Layout) con total control
fig.update_layout(
    title='Productividad del Programador (Versión Graph Objects)',
    xaxis_title='Días de la Semana',
    yaxis_title='Nivel de Energía / 100',
    template='plotly_dark' # Cambiamos el tema a modo oscuro para que sea más "pro"
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Configuración de datos
np.random.seed(1)
N = 100
random_x = np.linspace(0, 1, N)
random_y0 = np.random.randn(N) + 5
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N) - 5

fig = go.Figure()

# 1. Markers Custom (Cambiamos símbolo y tamaño)
fig.add_trace(go.Scatter(
    x=random_x, y=random_y0,
    mode='markers',
    name='Puntos Pro',
    marker=dict(size=10, color='LightSeaGreen', symbol='diamond')
))

# 2. Lines + Markers Custom (Cambiamos el tipo de línea a punteada)
fig.add_trace(go.Scatter(
    x=random_x, y=random_y1,
    mode='lines+markers',
    name='Mixto',
    line=dict(color='FireBrick', width=2, dash='dashdot'),
    marker=dict(size=6, color='black')
))

# 3. Lines Custom (Línea suave/curva con relleno)
fig.add_trace(go.Scatter(
    x=random_x, y=random_y2,
    mode='lines',
    name='Línea Suave',
    line=dict(color='RoyalBlue', width=4, shape='spline'), # 'spline' suaviza la curva
    fill='tonexty' # Rellena el área
))

# --- AQUÍ ESTÁ EL PODER DE LA PERSONALIZACIÓN ---
fig.update_layout(
    title=dict(
        text="Gráfico Altamente Personalizado",
        font=dict(size=24, color="DarkSlateGray"),
        x=0.5 # Centra el título
    ),
    xaxis=dict(title="Eje X (Tiempo)", showgrid=True, gridcolor='LightGray'),
    yaxis=dict(title="Eje Y (Valores)", showgrid=True, gridcolor='LightGray'),
    plot_bgcolor='white', # Fondo del área de trazado
    paper_bgcolor='#f8f9fa', # Fondo de toda la imagen
    legend=dict(
        yanchor="top",
        y=0.99,
        xanchor="left",
        x=0.01,
        bgcolor="rgba(255, 255, 255, 0.5)"
    ),
    hovermode="x unified" # Muestra todos los valores al pasar el mouse por el eje X
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px
import requests

def get_data_api(url):
    response = requests.get(url)
    data = response.json()

    df = pd.json_normalize(data)
    
    df = df[["name.common", "region", "population", "area", "capital", "cca3"]].rename(
        columns={"name.common": "country"}
    )
    return df
```

```python
import pandas as pd
import plotly.express as px
import streamlit as st


def plot_country_map(df, region):
    # Diccionario de scopes para el zoom de Plotly
    scope_map = {
        "Americas": "world",
        "Europe": "europe",
        "Asia": "asia",
        "Africa": "africa",
        "Oceania": "world",
    }

    current_scope = scope_map.get(region, "world")

    # 1. Crear el mapa base (Choropleth)
    fig = px.choropleth(
        df,
        locations="cca3",
        color="population",
        hover_name="country",
        hover_data=["area", "capital"],
        color_continuous_scale="Viridis",
        labels={"population": "Población"},
        title=f"Distribución de Población y Puntos Manuales en: {region}",
        scope=current_scope,
    )

    # 2. CREAR PUNTOS MANUALES
    data_puntos = {
        "lat": [19.4326, 40.4168, -34.6037, 52.5200, 35.6762, 30.0444],
        "lon": [-99.1332, -3.7038, -58.3816, 13.4050, 139.6503, 31.2357],
        "nombre": [
            "CDMX",
            "Madrid",
            "Buenos Aires",
            "Berlín",
            "Tokio",
            "El Cairo (Egipto)",
        ],
    }
    df_puntos = pd.DataFrame(data_puntos)

    # 3. AÑADIR LOS PUNTOS AL MAPA (ScatterGeo)
    fig.add_scattergeo(
        lat=df_puntos["lat"],
        lon=df_puntos["lon"],
        text=df_puntos["nombre"],
        mode="markers",
        marker=dict(
            size=12,
            color="red",
            symbol="star",  # <--- CAMBIADO: 'marker' es el pin estándar de Plotly
            line=dict(width=1, color="white"),
        ),
        name="Puntos de Interés",
    )

    fig.update_layout(
        margin={"r": 0, "t": 50, "l": 0, "b": 0},
        coloraxis_colorbar=dict(title="Habitantes"),
    )

    # Nota: Si estás en Jupyter usa fig.show(), en Streamlit st.plotly_chart
    st.plotly_chart(fig, width="stretch")
```

```python
# API URL
api_url = "https://restcountries.com/v3.1/all?fields=name,region,population,area,capital,cca3"

# Load data
df_api = get_data_api(api_url)

# In Jupyter, we just pick a region manually (e.g., 'Americas', 'Europe', 'Asia', 'Africa', 'Oceania')
selected_region = "Americas" 

# Filter the dataframe
df_api_filtered = df_api[df_api["region"] == selected_region]

# Execute the plot
plot_country_map(df_api_filtered, selected_region)
```
